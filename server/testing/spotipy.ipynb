```python
def get_artist_uri(artist_name):
    results = spotify.search(q=f'artist:{artist_name}', type='artist', limit=1)

    if results['artists']['items']:
        artist_uri = results['artists']['items'][0]['uri']
        return artist_uri
    else:
        print(f"Artist '{artist_name}' not found on Spotify.")
        return None
```

```python
def get_all_tracks(limit=50, offset=0):
    all_tracks = []

    # Initial request to get the total number of tracks
    results = spotify.search(q='a', type='track', limit=1, offset=0)
    total_tracks = results['tracks']['total']

    # Retrieve tracks in chunks (default chunk size is 50)
    while offset < total_tracks:
        results = spotify.search(q='a', type='track', limit=limit, offset=offset)
        all_tracks.extend(results['tracks']['items'])
        offset += limit

    return all_tracks

def create_tracks_dataframe(tracks):
    # Create a DataFrame with track information and audio features
    track_data = []
    for track in tracks:
        track_info = {
            'Track Name': track['name'],
            'Artist': track['artists'][0]['name'],
            'Track URL': track['external_urls']['spotify'],
        }
        # Get audio features for each track
        audio_features = spotify.audio_features([track['id']])
        if audio_features:
            track_info.update(audio_features[0])
        track_data.append(track_info)

    df = pd.DataFrame(track_data)
    return df

# Example: Get all tracks and create DataFrame
all_tracks = get_all_tracks()
all_tracks_df = create_tracks_dataframe(all_tracks)
print(all_tracks_df.columns.values)

# Display the DataFrame
print(all_tracks_df)
```
